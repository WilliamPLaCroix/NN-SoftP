```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# frozen llama with mlp head
conf_matrix_data_1 = np.array([[113, 40, 46, 45, 5, 1],
                                [62, 77, 81, 39, 6, 2],
                                [31, 54, 97, 64, 3, 0],
                                [28, 35, 73, 73, 1, 1],
                                [66, 58, 41, 37, 12, 0],
                                [55, 11, 9, 7, 3, 7]])
# frozen llama with lstm head
conf_matrix_data_2 = np.array([[20, 230, 0, 0, 0, 0],
                                [7, 260, 0, 0, 0, 0],
                                [8, 241, 0, 0, 0, 0],
                                [15, 196, 0, 0, 0, 0],
                                [7, 207, 0, 0, 0, 0],
                                [2, 90, 0, 0, 0, 0]])
# frozen llama with cnn head
conf_matrix_data_3 = np.array([[108, 69, 37, 17, 12, 7],
                                [76, 101, 56, 13, 17, 4],
                                [43, 84, 88, 26, 5, 3],
                                [37, 73, 74, 26, 0, 1],
                                [76, 72, 31, 15, 15, 5],
                                [47, 15, 10, 1, 8, 11]])

label_mapping = {
    3: 'True',
    2: 'Mostly True',
    1: 'Half True',
    4: 'Barely True',
    0: 'False',
    5: 'Pants on Fire'
}

for i, conf_matrix_data in enumerate([conf_matrix_data_1, conf_matrix_data_2, conf_matrix_data_3], start=1):
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix_data, annot=True, cmap="YlGnBu", fmt="d", linewidths=.5)
    plt.title(f'{["MLP", "LSTM", "CNN"][i-1]} Head - Frozen Llama 2')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.xticks(ticks=np.arange(6) + 0.5, labels=[label_mapping[i] for i in range(6)], rotation=45)
    plt.yticks(ticks=np.arange(6) + 0.5, labels=[label_mapping[i] for i in range(6)], rotation=45)
    plt.tight_layout()
    plt.show()
```
